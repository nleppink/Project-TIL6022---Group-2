# Modify LAX weather 

In this file the LAX weather is modified for usage for this research.

```python
import pandas as pd
import numpy as np
weather_LAX = pd.read_csv('LAX.csv')

#Clear the weather data by deleting non necessary columns
weather_LAX = weather_LAX.drop(['tmpf', 'dwpf', 'relh', 'drct', 'mslp', 'skyl1',
                                "skyl2",  'skyl3', 'skyl4',  'wxcodes', 'feel',
                                'ice_accretion_1hr', 'ice_accretion_3hr',
                                'ice_accretion_6hr', 'peak_wind_gust',
                                'peak_wind_drct', 'peak_wind_time', 'snowdepth',
                                'metar'], axis=1)

#Rename columns
weather_LAX = weather_LAX[['station', 'valid', 'sknt', 'gust',
                           'p01i', 'alti', 'vsby', 'skyc1',
                           'skyc2', 'skyc3', 'skyc4']]

weather_LAX.rename(columns={
    'sknt': 'Wind Speed [in knots]',
    'gust': 'Wind Gust [in knots]',
    'vsby': 'Visibility [in miles]',
    'p01i': 'One hour precipitation [in inches]',
    'alti': 'Pressure altimeter [in inches]',
    'skyc1': 'Sky Level 1 Coverage',
    'skyc2': 'Sky Level 2 Coverage',
    'skyc3': 'Sky Level 3 Coverage',
    'skyc4': 'Sky Level 4 Coverage'
}, inplace = True)

#Group the weather data with each data point for a measurement with a specific hour, ending with :53
weather_LAX = weather_LAX[weather_LAX['valid'].str.endswith(':53')]

#Rename the names for the table 
coverage_mapping = {
    'CLR': 1,  # Clear
    'FEW': 2,  # Few
    'SCT': 3,  # Scattered
    'BKN': 4,  # Broken
    'OVC': 5,  # Overcast
    'VV ': 6    # Vertical Visibility
}

weather_LAX['Sky Level 1 Coverage'] = weather_LAX['Sky Level 1 Coverage'].map(coverage_mapping)
weather_LAX['Sky Level 2 Coverage'] = weather_LAX['Sky Level 2 Coverage'].map(coverage_mapping)
weather_LAX['Sky Level 3 Coverage'] = weather_LAX['Sky Level 3 Coverage'].map(coverage_mapping)
weather_LAX['Sky Level 4 Coverage'] = weather_LAX['Sky Level 4 Coverage'].map(coverage_mapping)

display(weather_LAX)
```

```python
weather_LAX.to_csv('modified_weather_LAX.csv', index = False)
#Rename index
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Make sure that the 'valid' column has already been converted to datetime.
weather_LAX['valid'] = pd.to_datetime(weather_LAX['valid'])

#Set 'valid' as the index for easy time-based filtering.
weather_LAX = weather_LAX.set_index('valid')

#Filter the entire DataFrame (including Wind Speed and other columns) for the month of January 2013.
weather_LAX_january = weather_LAX['2013-01-01 00:53':'2013-01-31 23:53']
weather_LAX_februari = weather_LAX['2013-02-01 00:53':'2013-02-28 23:53']
weather_LAX_march = weather_LAX['2013-03-01 00:53':'2013-03-31 23:53']
weather_LAX_april = weather_LAX['2013-04-01 00:53':'2013-04-30 23:53']
weather_LAX_may = weather_LAX['2013-05-01 00:53':'2013-05-31 23:53']
weather_LAX_june = weather_LAX['2013-06-01 00:53':'2013-06-30 23:53']
weather_LAX_july = weather_LAX['2013-07-01 00:53':'2013-07-31 23:53']
weather_LAX_august = weather_LAX['2013-08-01 00:53':'2013-08-30 23:53']
weather_LAX_september = weather_LAX['2013-09-01 00:53':'2013-09-30 23:53']
weather_LAX_october = weather_LAX['2013-10-01 00:53':'2013-10-31 23:53']
weather_LAX_november = weather_LAX['2013-11-01 00:53':'2013-11-30 23:53']
weather_LAX_december = weather_LAX['2013-12-01 00:53':'2013-12-31 23:53']

weather_LAX_month = [weather_LAX_january, weather_LAX_februari,
                     weather_LAX_march, weather_LAX_april,
                     weather_LAX_may, weather_LAX_june,
                     weather_LAX_july, weather_LAX_august,
                     weather_LAX_september, weather_LAX_october,
                     weather_LAX_november, weather_LAX_december]

month_names = ['January', 'February', 'March', 'April', 'May', 'June',
                'July', 'August', 'September', 'October', 'November', 
               'December']

#Display(weather_LAX_january)

for i in range(len(weather_LAX_month)):
    plt.figure(figsize = (25, 5))
    plt.plot(weather_LAX_month[i]['Wind Speed [in knots]'])
    plt.gca().xaxis.set_major_locator(mdates.DayLocator())  
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d')) 
    plt.xticks(rotation=45, ha='right')

    # Add labels and titles
    plt.xlabel('Date')
    plt.ylabel('Wind Speed [knots]')
    plt.title(f'Wind Speed per Hour - {month_names[i]} 2013');
```
